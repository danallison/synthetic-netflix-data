```python
import pyspark
import pandas as pd
```

```python
spark = pyspark.sql.SparkSession.builder.getOrCreate()
```

```python
ratings = spark.read.csv('data/ratings.csv',
                        sep=',',
                        schema='movie_id INT, user_id INT, rating DOUBLE, date STRING',
                        header=False)
```

```python
ratings.printSchema()
```

```python
train, test = ratings.randomSplit([0.7, 0.3], seed=42)
```

```python
from pyspark.ml.recommendation import ALS

als = ALS(userCol='user_id',
          itemCol='movie_id',
          ratingCol='rating',
          rank=4,
          seed=42)

als_model = als.fit(train)
als_pred = als_model.transform(test)

pd.DataFrame(als_pred.take(3), columns=als_pred.columns)
```
